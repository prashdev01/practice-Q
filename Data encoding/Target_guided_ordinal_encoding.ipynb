# **Target guided ordinal encoding**

### It is technique used to encode categorical variable based on the relationship to the target variable. this encoding technique useful when we categorical vareible with large number of unique categories, and we want to use this variable as a feature in machine learning model.

### In the target Guided oridinal encoding we replace each catogery  in the categorcal variable with the  numerical variable based on the mean and meadian of the target variable for that category. this creates a monotonic relationship between the categorical varible and the target variable, which can improve the predictive power of model

```python
import pandas as pd

df=pd.DataFrame({
    'city': ['mumbai','umarkhed','mumbai','pune','nagpur'],
    'price' :[120,230,150,330,400]
})
```

```python
df
```

```python
price_mean = df.groupby('city')['price'].mean().to_dict()
```

```python
price_mean
```

```python
df['encoded_price'] = df['city'].map(price_mean)
```

```python
df
```

```python
import seaborn as sns

df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bill_according_to_day = df.groupby('total_bill')['size'].mean().to_dict()
total_bill_according_to_day
```

```python
df['total_bill_as_size'] = df['total_bill'].map(total_bill_according_to_day)
```
