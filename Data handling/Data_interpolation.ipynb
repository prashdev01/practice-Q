## **Linear ineterpolation** 

```python
import numpy as np
```

```python
x = np.array([1,3,5,7])
y = np.array([2,4,6,8])
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.interp(x_new,x,y)
```

```python
plt.scatter(x_new,y_interp)
```

# **Cubic interpolation**

```python
import numpy as np 
```

```python
x = np.array([1,2,3,4,5])
y = np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d

f= interp1d(x,y,kind='cubic')
```

```python
x_new = np.linspace(1,5,10)
y_interp= f(x_new)
print(y_interp)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```

## **Polynomial Interpolation**

```python
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,4,9,16,25])
```

```python
p = np.polyfit(x,y,4)
```

```python
x_new = np.linspace(1,5,10)
y_interp = np.polyval(p,x_new)
```

```python
plt.scatter(x,y)
```

```python
plt.scatter(x_new,y_interp)
```
