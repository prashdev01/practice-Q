```python
import seaborn as sns

# Step-1 Load dataset
df = sns.load_dataset('tips')
df.head()
```

```python
# perform one hot encoder
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
```

```python
# Step-2 splt the dataset
x = df[df.columns[:5]]
```

```python
x_encode = encoder.fit_transform(df[['total_bill', 'tip', 'sex', 'smoker', 'day']])
```

```python
import pandas as pd

x_encode_df = pd.DataFrame(x_encode.toarray(),columns=encoder.get_feature_names_out())
x = x_encode_df
```

```python
# y_encoded = encoder.fit_transform(df[['time']])
# y_encoded_df = pd.DataFrame(y_encoded.toarray(),columns=encoder.get_feature_names_out())
# y = y_encoded_df
y = df['time']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```
