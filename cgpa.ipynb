```python
def calculate_cgpa(results, out_of_marks):
    cgpa_results = {}
    for semester, total_marks in results.items():
        total_out_of_marks = out_of_marks.get(semester, 0)  # Get the corresponding out-of-marks
        cgpa = (total_marks / total_out_of_marks) * 4
        cgpa_results[semester] = cgpa
    return cgpa_results


# Given overall results and out-of-marks
results = {
    "sem-1,2": 556,
    "sem-3,4": 847,
    "sem-4,5": 657,
}

out_of_marks = {
    "sem-1,2": 1000,
    "sem-3,4": 900,
    "sem-4,5": 900,
}

# Calculate CGPA for each semester
cgpa_results = calculate_cgpa(results, out_of_marks)

# Display the CGPA for each semester
for semester, cgpa in cgpa_results.items():
    print(f"{semester} CGPA: {cgpa:.2f}")
```

```python
def calculate_overall_cgpa(results, out_of_marks, total_credit_hours):
    total_marks_obtained = sum(results.values())
    total_out_of_marks = sum(out_of_marks.values())
    
    cgpa = (total_marks_obtained / total_out_of_marks) * 4.0
    overall_cgpa = cgpa * (total_credit_hours / len(results))

    return overall_cgpa


# Given overall results and out-of-marks
results = {
    "sem-1,2": 556,
    "sem-3,4": 847,
    "sem-4,5": 657,
}

out_of_marks = {
    "sem-1,2": 1000,
    "sem-3,4": 900,
    "sem-4,5": 900,
}

# Assuming all subjects carry the same credit hours
total_credit_hours = 150  # Replace this with your actual total credit hours for graduation

# Calculate overall CGPA for graduation
overall_cgpa = calculate_overall_cgpa(results, out_of_marks, total_credit_hours)

# Display the overall CGPA
print("Overall CGPA: {:.2f}".format(overall_cgpa))
```

```python
def cgpa_to_percentage(cgpa):
    percentage = (cgpa / 4.0) * 100
    return percentage


# Given overall CGPA
overall_cgpa = 3.42  # Replace this with your actual overall CGPA

# Convert CGPA to percentage
percentage = cgpa_to_percentage(overall_cgpa)

# Display the percentage
print("Overall Percentage: {:.2f}%".format(percentage))
```
